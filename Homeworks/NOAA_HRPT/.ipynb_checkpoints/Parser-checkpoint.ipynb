```python
# Read data file
name = "20180710_165432_NOAA18"
# name = "20180715_015444_NOAA19"
# name = "20180717_125905_NOAA19"
with open("data_files/" + name + ".raw16", 'rb') as file:
        data = file.read()
```

```python
import struct
import numpy as np

# unpack binary data
length = int(len(data)/2)
print("File length:", length)

arr = np.zeros(length, dtype=np.uint16)

for i in range(length):
    arr[i] = struct.unpack('<h', data[2*i: 2*i + 2])[0]
```

```python
# Define initial sequence
seq = np.array([644, 367, 860, 413, 527, 149], dtype = np.uint16)
```

```python
def search_sequence_numpy(arr,seq):
    """ Find sequence in an array using NumPy only.

    Parameters
    ----------    
    arr    : input 1D array
    seq    : input 1D array

    Output
    ------    
    Output : 1D Array of indices in the input array that satisfy the 
    matching of input sequence in the input array.
    In case of no match, an empty list is returned.
    """

    # Store sizes of input array and sequence
    Na, Nseq = arr.size, seq.size

    # Range of sequence
    r_seq = np.arange(Nseq)

    # Create a 2D array of sliding indices across the entire length of input array.
    # Match up with the input sequence & get the matching starting indices.
    M = (arr[np.arange(Na-Nseq+1)[:,None] + r_seq] == seq).all(1)

    # Get the range of those indices as final output
    if M.any() >0:
        return np.where(np.convolve(M,np.ones((Nseq),dtype=int))>0)[0]
    else:
        return []         # No match found
```

```python
# Find all indices that match initial sequence
arr_of_indices = np.array(search_sequence_numpy(arr, seq))
arr_of_indices = arr_of_indices.reshape((int(arr_of_indices.shape[0]/(seq.shape[0])), seq.shape[0]))
```

```python
# Extract data for all channels
num_of_pix = int(10240/5)

channels = np.zeros((num_of_pix, arr_of_indices.shape[0], 5), dtype=np.uint16)

for k in range(arr_of_indices.shape[0]):
    temp = arr[arr_of_indices[k][0] + 750 - 1: arr_of_indices[k][0] + 10990 - 1]
    for i in range(num_of_pix):
        for j in range(5):
            channels[i][k][j] = temp[5 * i + j]
```

```python
# Choose R G B channels
R = channels[:, :, 1] >> 2
G = channels[:, :, 2] >> 2
# B = channels[:, :, 4] >> 2
# Inversed B
B = (2**10 - channels[:, :, 4]) >> 2
```

```python
from matplotlib import pyplot as plt

# Function to plot histogram
def plot_hist(array):
    hist,bins = np.histogram(array.flatten(),256,[0,256])

    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max()/ cdf.max()

    plt.plot(cdf_normalized, color = 'b')
    plt.hist(array.flatten(),256,[0,256], color = 'r')
    plt.xlim([0,256])
    plt.legend(('cdf','histogram'), loc = 'upper left')
    plt.grid()

    plt.show()
    return
```

```python
# Channel R histogram
plot_hist(R)
```

```python
# Channel G histogram
plot_hist(G)
```

```python
# Channel B histogram
plot_hist(B)
```

```python
from PIL import Image

# Export original non-equalized pseudo RGB
image = np.zeros(shape = (channels.shape[0], channels.shape[1], 3))
image[:, :, 0] = R
image[:, :, 1] = G
image[:, :, 2] = B

img = Image.fromarray(image.astype('uint8'), mode = "RGB")
img.save("output/original/" + name +"_original_.bmp")
```

```python
from skimage import exposure

# Parameters
clip_limit=0.1

# Histogram adaptive equalization
image[:, :, 0] = exposure.equalize_adapthist(R, clip_limit=clip_limit) * 255
image[:, :, 1] = exposure.equalize_adapthist(G, clip_limit=clip_limit) * 255
image[:, :, 2] = exposure.equalize_adapthist(B, clip_limit=clip_limit) * 255
```

```python
img = Image.fromarray(image.astype('uint8'), mode = "RGB")
img.save("output/equalized/" + name +"_eq_.bmp")
```
